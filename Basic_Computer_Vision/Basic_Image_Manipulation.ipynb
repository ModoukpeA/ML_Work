```python
import cv2
import numpy as np
import requests
from matplotlib import pyplot as plt
```

```python
# Specify the image URL
url = 'https://cdn.britannica.com/45/5645-050-B9EC0205/head-treasure-flower-disk-flowers-inflorescence-ray.jpg'  # Replace with your image URL

# Fetch the image from the URL
response = requests.get(url)
image_array = np.asarray(bytearray(response.content), dtype=np.uint8)
```

```python
# Decode the image array to an OpenCV image
image = cv2.imdecode(image_array, cv2.IMREAD_COLOR)

# Convert color to RGB (optional for matplotlib visualization)
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Grayscale
grayscale_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Apply histogram equalization
equalized_image = cv2.equalizeHist(grayscale_image)
```

```python
# Display the image
fig, axes = plt.subplots(1,4, figsize=(15,5))
axes[0].imshow(image)
axes[0].set_title('Image from URL')
axes[0].axis('off')

axes[1].imshow(image_rgb)
axes[1].set_title('RGB Image')
axes[1].axis('off')

axes[2].imshow(grayscale_image, cmap='gray')
axes[2].set_title('Grayscale Image')
axes[2].axis('off')

axes[3].imshow(equalized_image, cmap='gray')
axes[3].set_title('Equalized Image')
axes[3].axis('off')
plt.show()
```

```python
# Resize the image to 128x128 pixels
resized_image = cv2.resize(image_rgb, (128, 128))

plt.imshow(resized_image, cmap='gray')
plt.title('Resized Image (128x128)')
plt.show()
```

```python
# Apply Gaussian blur
blurred_image = cv2.GaussianBlur(grayscale_image, (51, 51), 0)

plt.imshow(blurred_image, cmap='gray')
plt.title('Gaussian Blurred Image')
plt.show()
```

```python
# Apply Canny edge detection
edges = cv2.Canny(grayscale_image, 100, 150)

plt.imshow(edges, cmap='gray')
plt.title('Canny Edge Detection')
plt.show()
```
