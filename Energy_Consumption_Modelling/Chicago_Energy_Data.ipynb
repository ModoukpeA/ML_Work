# Importing Required Libraries

```python
# Important Libraries

import pandas as pd
```

# Data Preparation and Preprocessing


```python
# Loading the datasest
energy_usage = pd.read_csv("Energy_Usage_2010.csv")
```

```python
# printing the dataset
energy_usage
```

```python
energy_usage.shape
```

The energy usage data set has 67051 rows and 73 columns

```python
# short summary of the dataset
energy_usage.info()
```

```python
df1 = energy_usage.iloc[:,0:19]
df1
```

```python
df2 = energy_usage.iloc[:,33:49]
# dropping the therms total sqft column
df2 = df2.drop(["THERMS TOTAL SQFT"], axis=1)
df2
```

```python
df3 = energy_usage.iloc[:,63:]
df3
```

```python
# merging the electricity dfs
electricity_data = pd.concat([df1, df2, df3], axis=1)
electricity_data 
```

```python
electricity_data.shape
```

```python
# summary info of the electricty data
electricity_data.info()
```

```python
# checking for null values per column
electricity_data.isnull().sum()
```

Selecting only columns that are not of decriptive statistics


```python
new_df1 = electricity_data.iloc[:, 0:19]
new_df1
```

```python
new_df2 = electricity_data.iloc[:, 34:]
new_df2
```

```python
# merging the electricity dfs
final_df = pd.concat([new_df1, new_df2], axis=1)
final_df 
```

```python
final_df.info()
```

```python
final_df= final_df.drop(["RENTER-OCCUPIED HOUSING PERCENTAGE","OCCUPIED UNITS PERCENTAGE"], axis=1)
final_df
```

```python
final_df.info()
```

```python
final_df.isnull().sum()
```

```python
# filling nan values
final_df = final_df.interpolate(method ='linear')

final_df.info()
```

```python
final_df.isnull().sum()
```

```python
final_df.shape
```

```python
final_df.head()
```

```python
final_df["BUILDING TYPE"].unique()
```

```python
# dropping the nan values
final_df = final_df.dropna()
```

```python
final_df.info()
```
